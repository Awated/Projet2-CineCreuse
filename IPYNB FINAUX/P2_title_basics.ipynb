```python
import pandas as pd
import numpy as np
from google.colab import files
```

```python
# df = pd.read_csv('title.basics.tsv.gz', sep="\t")# si vous avez telechargez le fichier dans votre ordi
df = pd.read_csv('https://datasets.imdbws.com/title.basics.tsv.gz', sep= "\t") #si vous n'avez pas ou ne voulez pas telecharger les datasets


# relancer tout pour telecharger le resultat final en  csv
```

```python
df.head()
```

```python
display(df.info())
print(df.shape)
```

```python
df = df.replace('\\N',np.nan) # fonction pour modifier toutes les valeurs \N par des NaN
```

```python
df.shape
```

```python
print(df.shape)
```

```python
# suppression des NaN de la colonne runtimeMinutes
df['runtimeMinutes'] = df['runtimeMinutes'].dropna()
```

```python
# condition filtrant sur tout public (non Adult)
df= df[df['isAdult'].isin([0,'0'])]

```

```python
# condition filtrant sur 5 de nos critères : court-métrage(short / tvshort), long metrage(movie/tv_movie/tvSpecial)
condition = df['titleType'].isin(["short","movie","tvMovie","tvShort","tvSpecial, video"])

df_final = df[condition]
print(df_final.shape)
```

```python
# suppression des colonnes Endyear car quasi totalité de NaN et primaryTitle car quasi doublon de Original Title
df_final = df_final.drop(columns= ['endYear', 'primaryTitle'], axis=1)
```

```python
df_final = df_final.drop(columns= ['isAdult'])
```

```python
print(df_final.shape)
```

```python
check_nan = df_final.isna().sum()
print(check_nan)
```

```python
df_final = df_final.dropna()
```

```python
print(df_final.shape)
```

```python
# fonction pour conserver les lignes avec les genres selectionnés
def genre_selected(data):
  condition = data.genres.str.contains('Comedy|Action|Sci-Fi|Animation')    # genres conservés basés sur notre etude de marché
  return data[condition]
```

```python
df_final = genre_selected(df_final)
```

```python
print(df_final.shape)
```

```python
# pas de doublons
df_final.drop_duplicates(subset = "tconst").shape
```

```python
df_final[[ 'startYear','runtimeMinutes']] = df_final[['startYear','runtimeMinutes']].astype("int32")
```

```python
# new['runtimeMinutes'].max() pour verifier qu'il s'arrete bien a 1h45min
df_final= df_final[df_final['runtimeMinutes'].isin( range ( 1,106))]
```

```python
check_nan =df_final.isna().sum()
print(check_nan)
```

```python
df_final.info()
```

```python
df_final = df_final[df_final['startYear'] >= 1970]
```
