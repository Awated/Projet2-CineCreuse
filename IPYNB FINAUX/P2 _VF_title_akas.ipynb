```python
import numpy as np
```

```python
df1 = pd.read_csv('https://datasets.imdbws.com/title.akas.tsv.gz', sep = '\t')
display(df1)
```

```python
def trans_nan(dataframe):
    return dataframe.replace('\\N',np.nan)
df1 = trans_nan(df1)
```

```python
# compter le nombre de fois que chaque veleur unique reviens dans la colonne region
count_ship = df1['region'].value_counts(dropna=False)
print(count_ship)
```
