```python
# library à importer
import csv
import pandas as pd
import numpy as np
```

```python
#import et transformaton du dataset en dataframe
actors = pd.read_csv("https://datasets.imdbws.com/name.basics.tsv.gz", sep="\t")# Cette case prend un peu de temps c'est normal le temps d'execution
```

```python
df = actors
```

```python
df.head(20)
```

```python
# Modification de tous les str les \N par des nan
df = df.replace('\\N',np.nan)
```

```python
check_size = print(df.shape)
```

```python
# Convert the 'deathYear' column to a nullable integer type exemple: <NaN>
# This allows NaN values to be represented in the integer column
df['deathYear'] = df['deathYear'].astype('Int64')
```

```python
# fonction pour conserver les lignes avec les genres selectionnés
def profession_selected(data):
  condition = data.primaryProfession.str.contains('actor|actress|director-Fi|producer')    # genres conservés basés sur notre etude de marché
  return data[condition]
```

```python
df = profession_selected(df)
```

```python
def profession(row):
  # Check if the value in 'primaryProfession' is a string with the built in function is instance(value to check , type to check)
  if isinstance(row['primaryProfession'], str):
    if 'actor' in row['primaryProfession']:
      return row['primaryProfession']
    elif 'actress' in row['primaryProfession']:
      return row['primaryProfession']
    elif 'director' in row['primaryProfession']:
      return row['primaryProfession']
    elif 'producer' in row['primaryProfession']:
      return row['primaryProfession']
    else:
      # Return None if the value is not in ['actor',actress,director,producer]
      return None
  else:
    # Return None if the value is not a string (e.g., NaN)
    return None

df['profession'] = df.apply(profession,axis=1)
# Cette case prend un peu de temps c'est normal le temps d'execution
```

```python
df = df.dropna(subset= ['profession'])
```

```python
def mort_avant(row):
    # Check if the value in 'deathYear' is not NaN before performing the comparison
    if pd.notna(row['deathYear']):
        if row['deathYear'] < 1970:
            return row['deathYear']
        else:
            return None
    else:
        # Return None if the value is NaN
        return None

df['death'] = df.apply(mort_avant,axis=1)
```

```python
df['death'] = df['death'].astype('Int64') #changer en Int64 for adapt nan
```
