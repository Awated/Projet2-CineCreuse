```python
import pandas as pd
df = pd.read_csv("https://datasets.imdbws.com/title.ratings.tsv.gz",sep= "\t")
```

```python
# recherche de la moyenne de averageRating
print(df['averageRating'].mean())
```

```python
# moyenne du npmbre de votes sur l'ensemble du dataset
print(df['numVotes'].mean())
```

```python
# fonction de filtre du dataset pour tous les films dont le average rating est supérieur à 7 ET nomber de votes > 10 000
def clean_dataset(datasets):
  condition = (datasets['averageRating'] > 7) & (datasets['numVotes'] > 10000)
  datasets = datasets[condition]
  return datasets


ratings = clean_dataset(df)
```

```python
# afficher les valeurs triées par averageRating
check_top_rate = ratings['averageRating'].sort_values(ascending= False)
display(check_top_rate)
```

```python
# renommer les colonnes
ratings= ratings.rename(columns= {"averageRating":"moyenneVote",
                                 "numVotes":"NombreVotes"})
```
